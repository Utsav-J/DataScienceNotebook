```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

# Stat Graph

```python
data = {
    'x' : [1,2,3,4,5],
    'y' : [2,4,6,8,9],
    'hue' : ['A','B','A','B','C']
}
# data = pd.DataFrame(data)
```

```python
sns.scatterplot(x='x',y='y',hue='hue',data=data)
plt.xlabel("Axis X")
plt.ylabel("Axis Y")
plt.show()
```

# Built in

```python
tips = sns.load_dataset('tips')
sns.scatterplot(x="total_bill", y = "tip",hue='sex', data = tips)
plt.legend(loc = "upper center")
plt.show()
```

# Joint Plots

```python
data = {
    'x' : [1,2,3,4,5],
    'y' : [2,2,5,4,5],
    'hue': ['a','b','b','a','b']
}
```

```python
sns.jointplot(x='x', y='y', data = data, kind = 'scatter')
```

```python
tips = sns.load_dataset('tips')
sns.jointplot(x = 'total_bill', y = 'tip', hue = "sex", data = tips, kind = "scatter")
plt.show()
```

```python
sns.jointplot(x='x',y='y',hue='hue',data=data)
plt.show()
```

# Annotations

```python
x = [1, 2, 3, 4, 5]
y = [2, 4, 3, 5, 1]
sns.scatterplot(x=x,y=y)

for i in range(5):
  plt.annotate(f"{x[i]},{y[i]}",xy=(x[i],y[i]),xytext=(x[i]-0.1,y[i]+0.1))
  # plt.text(x[i],y[i],f"{x[i]},{y[i]}",ha='center',va='bottom')

plt.show()
```

# 3D Plots

```python
import matplotlib.pyplot as plt
import seaborn as sns
from mpl_toolkits.mplot3d import Axes3D

x = [1, 2, 3, 4, 5]
y = [2, 4, 3, 5, 1]
z = [10, 15, 20, 25, 30]
```

```python
fig = plt.figure()
ax = fig.add_subplot(111,projection = '3d')
scatter = ax.scatter(x,y,z) ######################
ax.set_xlabel('X Axis')
ax.set_ylabel('Y Axis')
ax.set_zlabel('Z Axis')
ax.view_init(elev=20, azim=45)
plt.grid(False)
plt.show()
```

```python
figure = plt.figure()
axs = figure.add_subplot(111,projection = '3d')
lineplot = axs.plot3D(x,y,z) ######################
axs.set_xlabel("X Axis")
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
newx = np.linspace(0,5,10)
newy = np.linspace(0,5,10)
x,y = np.meshgrid(newx,newy)

def f(x,y):
  return x**2+y**2

z = f(x,y)
fig = plt.figure()
axs = fig.add_subplot(111,projection='3d')

surface = axs.plot_surface(x,y,z, cmap='viridis', linewidth = 0, antialiased=True) ####################
axs.view_init(elev=23)
plt.show()
```
