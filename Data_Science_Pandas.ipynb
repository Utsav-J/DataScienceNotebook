# Series

One dimensional special data structure that is capable of holding data of various types and makes use of unique indexing for record wise unique access.

They are the building blocks for a DataFrame

## From list

```python
lst = [1,2,3,4,5]
ser = pd.Series(lst)
print(ser)
print(ser[0])
print(ser.loc[0])
print(ser.iloc[0])
```

## From dict

```python
dic = {'a':1,'b':2,'c':3}
s1 = pd.Series(dic)
print(s1)
print(s1['a'])
print(s1.loc['a'])
print(s1.iloc[0])
```

## Ranking

```python
df = pd.DataFrame(data={'Animal': ['fox', 'Kangaroo','deer','spider', 'snake'],'Number_legs': [4, 2, 4, 8, np.nan]})
df
```

```python
df['default_ranks'] = df['Number_legs'].rank()
df
```

```python
df['max_rank'] = df['Number_legs'].rank(method = 'max')
df
```

```python
df["pct_rank"] = df["Number_legs"].rank(pct= True)
df
```

```python
df['na-option'] = df["Number_legs"].rank(na_option='bottom')
df
```

```python
df['descending_rank'] = df["Number_legs"].rank(ascending = False, method = 'min', na_option = 'bottom')
df
```

```python
sorted_df = df.sort_values(by = "descending_rank")
sorted_df
```

# Country Sorting

```python
age_list = [['Afghanistan', 1952, 8425333, 'Asia'],
 ['Australia', 1957, 9712569, 'Oceania'],
 ['Brazil', 1962, 76039390, 'Americas'],
 ['China', 1957, 637408000, 'Asia'],
 ['France', 1957, 44310863, 'Europe'],
 ['India', 1952, 3.72e+08, 'Asia'],
 ['United States', 1957, 171984000, 'Americas']]

df = pd.DataFrame(age_list, columns = ["Name","Year","Population","Continent"])
df.head()
```

```python
df['age_rank'] = df['Year'].rank(method = 'min')
df
```

```python
df.sort_values(by="Population", ascending = False)
```
