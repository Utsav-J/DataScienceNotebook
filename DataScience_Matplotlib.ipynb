```python
import numpy as np
import matplotlib.pyplot as plt
```

# Subplots

```python
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])
plt.subplot(1,2,1)
plt.plot(x,y)

x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])
plt.subplot(1,2,2)
plt.plot(x,y)

plt.show()
```

```python
x = np.array([0, 1, 2, 3])
y = x**2
plt.subplot(2,3,1)
plt.plot(x,y)

x = np.array([0, 1, 2, 3])
y = x**3
plt.subplot(2,3,2)
plt.plot(x,y)

x = np.array([0, 1, 2, 3])
y = x**4
plt.subplot(2,3,3)
plt.plot(x,y)

x = np.array([0, 1, 2, 3])
y = x**5
plt.subplot(2,3,4)
plt.plot(x,y)

x = np.array([0, 1, 2, 3])
y = x**6
plt.subplot(2,3,5)
plt.plot(x,y)
x = np.array([0, 1, 2, 3])
y = x**6
plt.subplot(2,3,6)
plt.plot(x,y)
plt.show()
```

# Multiple Plots

```python
x = np.array([1,2,3])
y = x*2

fig, axes = plt.subplots(2,2)

axes[0,0].plot(x,y)
axes[0,1].scatter(x,y)
axes[1,0].bar(x,y)
axes[1,1].hist(x,y)
plt.show()
```

# Axes Control

```python
x = np.arange(1,10,1)
y = x*2 - 1

plt.plot(x,y)

plt.xlim(0,20)
plt.ylim(0,30)

plt.show()
```

# Ticks

```python
x = np.arange(1,10,1)
y = x*2 - 1

plt.xticks(x,['One','Two',"Three","Four","Five","Six","Seven","Eight","Nine"])

plt.xticks(rotation=45)

plt.plot(x,y)
plt.show()
```

# Labels and Titles

```python
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('Practicing Data Science', fontsize = 20)
plt.scatter(x,y)
plt.show()
```

# Legend

```python
plt.plot([1,2,3,4,5],[2,4,6,7,8],label="First line")
plt.plot([2,4,6,7,8],[1,2,3,4,5],label="Second line")
plt.scatter([2,4,5,7,8],[1,9,3,4,5],label="Third line")
plt.legend()
plt.show()
```

```python
plt.plot([1,2,3,4,5],[2,4,6,7,8],label="First line")
plt.plot([2,4,6,7,8],[1,2,3,4,5],label="Second line")
plt.legend(loc = 'lower left')
plt.show()
```

# Annotations

```python
text = "Look at this dood"
xy = (1,2)
xytext = (3,4)
xycoords='data'
textcoords='data'
arrowprops = dict(arrowstyle="-", facecolor='purple')
annotation_clip = True

plt.plot([1,2,3,4,5],[2,4,6,7,8],marker='p')
plt.annotate(text,xy,xytext,xycoords,textcoords,arrowprops,annotation_clip,fontsize=20, color='orange')
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title('Annotations bbgurl')
plt.show()
```

```python
plt.plot([1,2,3,4,5],[2,4,6,7,8],marker='p')
plt.annotate(
    'Here is this one',
    (3,6),
    (2,7),
    'data',
    'data',
    arrowprops = dict(arrowstyle = "-[", facecolor = "red"),
    bbox = dict(boxstyle= 'round, pad = 1'),
    color = 'white'
)
plt.show()
```

# Saving

```python
plt.plot([1,2,3,4,5],[2,4,6,7,8],marker='p')
plt.annotate("Here is this", (3,6),(2,7),xycoords = "data",textcoords = "data",bbox=dict(boxstyle='round,pad=0.5'), arrowprops=dict(arrowstyle='-[', facecolor='red'))
plt.show()
plt.savefig('first.pdf')
```
